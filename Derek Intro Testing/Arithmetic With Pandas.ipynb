# Arithmetic With Pandas

This notebook defines a simple arithmetic function and applies it to a pandas DataFrame.

```python
import pandas as pd


def add_numbers(a, b):
    return a + b
```

```python
df = pd.DataFrame(
    {
        "first_number": [1, 5, 10],
        "second_number": [2, 3, 4],
    }
)

df["sum"] = df.apply(
    lambda row: add_numbers(row["first_number"], row["second_number"]),
    axis=1,
)

df
```
